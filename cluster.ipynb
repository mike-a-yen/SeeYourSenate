```python
from app import app,db

import numpy as np
import pandas as pd

import nltk
from nltk.stem.snowball import SnowballStemmer
from textblob import TextBlob
import re
import os
import codecs

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans, DBSCAN

import matplotlib.pyplot as plt
import mpld3
%matplotlib inline

stemmer = SnowballStemmer('english')
stopwords = nltk.corpus.stopwords
```

```python
reviews = pd.read_sql_table('reviews',app.config['SQLALCHEMY_DATABASE_URI'])
```

```python
bid = reviews.sample().iloc[0]['business_id']
print(bid)
```

```python
revs = reviews[reviews['business_id']==bid]
len(revs)
```

```python
def tokenize_and_stem(text):
    tokens = [word for word in TextBlob(text).words]
    filtered_tokens = [re.sub('[^a-zA-Z0-9]','',w) for w in tokens]
    stems = [stemmer.stem(t) for t in filtered_tokens]
    return stems
```

```python
tfidf = TfidfVectorizer(max_df=0.9, max_features=20000,
                        min_df=0.1, stop_words='english',
                        use_idf=True, tokenizer=tokenize_and_stem, ngram_range=(1,3))
```

```python
tfidf_matrix = tfidf.fit_transform(revs['text'])
```

```python
km = KMeans(n_clusters=5)
km.fit(tfidf_matrix)
clusters = km.labels_.tolist()
```
